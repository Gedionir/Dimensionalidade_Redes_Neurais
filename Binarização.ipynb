**Binarização** - Convertendo imagens em tons de cinza para imagens preto e branco onde todos os pixels
possuem 0 ou 255 como valores de intensidade.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

```

```python
img = cv2.imread("fotos/foto_pb.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
suave = cv2.GaussianBlur(img, (7, 7), 0)
(T, bin) = cv2.threshold(suave, 161, 255, cv2.THRESH_BINARY)
(T, binI) = cv2.threshold(suave, 161, 255, cv2.THRESH_BINARY_INV)
resultado = np.vstack([
    np.hstack([suave, bin]),
    np.hstack([binI, cv2.bitwise_and(img, img, mask = binI)])
])
cv2_imshow(resultado)
cv2.waitKey(0)

```

```python
img = cv2.imread('fotos/olho-color.jpg')
cv2_imshow(img)
(cA, cVd, cV) = cv2.split(img)
cv2_imshow(cV)
cv2_imshow(cVd)
cv2_imshow(cA)
cv2.waitKey(0)
```

**Sistemas de Cores** - Colorido, tons de cinza, além de outros coloridos como o L*a*b* e o HSV.

```python
img = cv2.imread('fotos/foto_color.jpg')
cv2_imshow(img)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
hsv= cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow (hsv)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2_imshow(lab)
cv2.waitKey(0)
```
